## Question 2
Flights are often delayed. Consider the following questions exploring delay patterns.

(a) Which date has the largest average departure delay? Which date has the largest average arrival delay?

```python
flights_df['date'] = flights_df['year'].astype(str) + '/' + flights_df['month'].astype(str) + '/' + flights_df['day'].astype(str)

flights_df_agg = pd.DataFrame(flights_df.groupby('date').mean())
flights_df_agg.reset_index(level=0,inplace=True)

max_avg_arr_delay = max(flights_df_agg['arr_delay'])
max_avg_dep_delay = max(flights_df_agg['dep_delay'])

print("Largest arrival delay was on date:", flights_df_agg.loc[flights_df_agg['arr_delay'] == max_avg_arr_delay,'date'].item(),"with Arrival Delay:",max_avg_arr_delay)

print("Largest departure delay was on date:", flights_df_agg.loc[flights_df_agg['dep_delay'] == max_avg_dep_delay,'date'].item(),"with Departure Delay:", max_avg_dep_delay)
```

Maximum arrival and departure delay is on the same date: 2013/3/8 with average arrival delay of 85.86 and departure delay of 83.53.

(b) What was the worst day to fly out of NYC in 2013 if you dislike delayed flights?


```python
flights_df_agg['total_delay'] = (flights_df_agg['arr_delay'] + flights_df_agg['dep_delay'])/2
s = max(flights_df_agg['total_delay'])

print(flights_df_agg.loc[flights_df_agg['total_delay'] == s,'date'].item(),": date has the maximum average total delay:", s)
print(flights_df_agg.loc[flights_df_agg['dep_delay'] == max_avg_dep_delay,'date'].item(),": date has the maximum departure delay:", max_avg_dep_delay)

dep_delay= flights_df[flights_df['dep_delay']>0]
dep_delay = pd.DataFrame(dep_delay.groupby('date').count())
dep_delay.reset_index(level=0,inplace=True)
print(dep_delay.loc[dep_delay['Unnamed: 0'] == max(dep_delay['Unnamed: 0']),'date'].item(),": date with most number of delayed flights in a day -", max(dep_delay['Unnamed: 0']))

print(flights_df.loc[flights_df['dep_delay'] == max(flights_df['dep_delay']),'date'].item(),": date has maximum delay in departure of a flight, with delay of",max(flights_df['dep_delay']))
```

Worst day to fly out of NYC in 2013 if you disliked delayed flights would depend on, what someone considered as an inconvenience:

(c) Are there any seasonal patterns in departure delays for flights from NYC?

```python
plt.style.use('ggplot')

plt.plot(flights_df.groupby('month')['dep_delay'].mean())
plt.xlabel('Month')
plt.ylabel('Average departure delay in mins')
plt.title('Average depature delay(in mins) by month')
plt.show()

dep_delay = flights_df[flights_df['dep_delay'] > 0]
plt.plot(dep_delay.groupby('month').count())
plt.xlabel('Month')
plt.ylabel('Number of flights')
plt.title('Number of flights delayed every month')
plt.show()
```

As we can see from the plots above, there is seasonal pattern with departure delays with maximum number of flights getting delayed in the month of 'June', 'July' and 'August'. Also, the average depature delay is highest in these months. As this is a period of summer vacation, there is more traffic at the airports and hence the delay.
Additionally, during the winter season, there are less number of people at the airport, hence the delay is lesser and total number of flights delayed is also less.

## Question 6
What weather conditions are associated with flight delays leaving NYC? Use graphics to explore.

```python
weather_df = pd.read_csv('weather.csv.bz2')

weather_df = pd.DataFrame(weather_df.groupby(['year','month','day','origin']).mean())
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)
weather_df.reset_index(level = 0, inplace = True)

weather_df.head()
weather_df.shape
```

```python
flights_agg = flights_df.groupby(['year','month','day','origin']).mean()
flights_agg.reset_index(level = 0, inplace = True)
flights_agg.reset_index(level = 0, inplace = True)
flights_agg.reset_index(level = 0, inplace = True)
flights_agg.reset_index(level = 0, inplace = True)

flights_agg.shape
```

```python
flights_weather = pd.merge(weather_df,flights_agg,on=['year','month','day','origin'], how='inner')
flights_weather.shape
```

```python
rcParams['figure.figsize'] = 20, 20
fig = plt.figure()
ax1 = fig.add_subplot(331)
ax1.scatter(flights_weather['temp'], flights_weather['dep_delay'], c = 'red')
xlabel('Temp in farenheight')
ylabel('Avg dep delay in mins')
title('Temp Vs. Depature Delay')

ax2 = fig.add_subplot(332)
ax2.scatter(flights_weather['dewp'], flights_weather['dep_delay'], c = 'blue')
xlabel('Dewpoint')
ylabel('Avg dep delay in mins')
title('Dewpoint Vs. Depature Delay')

ax3 = fig.add_subplot(333)
ax3.scatter(flights_weather['humid'], flights_weather['dep_delay'], c = 'black')
xlabel('Humidity')
ylabel('Avg departure delay in mins')
title('Humidity Vs. Depature Delay')

ax4 = fig.add_subplot(334)
ax4.scatter(flights_weather['wind_gust'], flights_weather['dep_delay'], c = 'darkred')
xlabel('Wind gust')
ylabel('Avg departure delay in mins')
title('Wind gust Vs. Depature Delay')

ax5 = fig.add_subplot(335)
ax5.scatter(flights_weather['wind_dir'], flights_weather['dep_delay'], c = 'orange')
xlabel('Wind direction')
ylabel('Avg departure delay in mins')
title('Wind direction Vs. Depature Delay')

ax6 = fig.add_subplot(336)
ax6.scatter(flights_weather['wind_speed'], flights_weather['dep_delay'], c = 'darkgreen')
xlabel('Wind speed')
ylabel('Avg departure delay in mins')
title('Wind speed Vs. Depature Delay')

ax7 = fig.add_subplot(337)
ax7.scatter(flights_weather['precip'], flights_weather['dep_delay'], c = 'violet')
xlabel('Precip')
ylabel('Avg departure delay in mins')
title('Precip Vs. Depature Delay')

ax8 = fig.add_subplot(338)
ax8.scatter(flights_weather['pressure'], flights_weather['dep_delay'], c = 'green')
xlabel('Pressure')
ylabel('Avg departure delay in mins')
title('Pressure Vs. Depature Delay')

ax9 = fig.add_subplot(339)
ax9.scatter(flights_weather['visib'], flights_weather['dep_delay'], c = 'pink')
xlabel('Visibility')
ylabel('Avg departure delay in mins')
title('Visibility Vs. Depature Delay')
```

```python
mod1 = smf.ols(formula='dep_delay ~ temp', data=flights_weather).fit()
print ("Regression model summary for dep_delay and temperature:\n", mod1.params)

mod2 = smf.ols(formula='dep_delay ~ dewp', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and dewpoint:\n", mod2.params)

mod3 = smf.ols(formula='dep_delay ~ humid', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and humidity:\n", mod3.params)

mod4 = smf.ols(formula='dep_delay ~ wind_gust', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and wind_gust:\n", mod4.params)

mod5 = smf.ols(formula='dep_delay ~ wind_dir', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and wind_direction:\n", mod5.params)

mod6 = smf.ols(formula='dep_delay ~ wind_speed', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and wind_speed:\n", mod6.params)

mod7 = smf.ols(formula='dep_delay ~ precip', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and precipitation:\n", mod7.params)

mod8 = smf.ols(formula='dep_delay ~ pressure', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and pressure:\n", mod8.params)

mod9 = smf.ols(formula='dep_delay ~ visib', data=flights_weather).fit()
print ("\nRegression model summary for dep_delay and visibility:\n", mod9.params)
```

Based on the plots and regression models above, the analysis for effect of weather conditions on NYC flights in 2013 is as explained below:

1. Departure delay and temperature: With every one farenheight increase in temperature, there is 0.08 mins of increase in the departure delay.

2. Departure delay and dewpoint: With every one unit increase in dewpoint, the departure delay increases by 0.15 mins.

3. Departure delay and humidity: With every one unit increase in humidity, the departure delay increases by 0.38 mins.

4. Departure delay and wind gust: With every one unit increase in wind gust, the departure delay increase by 0.02 mins.

5. Departure delay and wind direction: With every one unit increase in wind direction, the departure delay decreases by 0.02 mins.

6. Departure delay and wind speed: With every one unit increase in wind speed, the departure delay increase by 0.03 mins.

7. Departure delay and precipitation: With every one unit increase in precipitation, the departure delay increase by 447.58 mins. This is a highest effect on the departure delay.

8. Departure delay and pressure: With every one unit increase in pressure, the departure delay decreases by 0.49 mins.

9. Departure delay and visibility: With every one unit increase in visibility, the departure delay decreases by 3.40 mins.

Based on the above analysis, it is inferred that precipitation, pressure, humidity and visibility has some significant impact on the departure delay times.

Based on the above graphics and regression results, we can say that weather does have an effect on flights from NYC in 2013. However, more understanding and statistical modelling is required to understand the significance level of these individual features and their impacts.
